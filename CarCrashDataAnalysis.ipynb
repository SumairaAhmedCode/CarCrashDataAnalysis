```python
# Data preprocessing

# Convert 'CRASH DATE' column to datetime
data['CRASH DATE'] = pd.to_datetime(data['CRASH DATE'])
```

```python
# Data visualization

# Visualize the number of crashes over time
plt.figure(figsize=(12, 6))
data['CRASH DATE'].value_counts().sort_index().plot()
plt.title('Number of Crashes Over Time')
plt.xlabel('Date')
plt.ylabel('Count')
plt.show()
```

```python

# Visualize the number of crashes in each borough
plt.figure(figsize=(10, 6))
sns.countplot(x='BOROUGH', data=df)
plt.title('Number of Crashes by Borough')
plt.xlabel('Borough')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()


```

```python
# Visualize the contributing factors for vehicle crashes
plt.figure(figsize=(12, 8))
sns.countplot(x='CONTRIBUTING FACTOR VEHICLE 1', data=data, order=data['CONTRIBUTING FACTOR VEHICLE 1'].value_counts().index)
plt.title('Contributing Factors for Vehicle Crashes')
plt.xlabel('Contributing Factor')
plt.ylabel('Count')
plt.xticks(rotation=90)
plt.show()

```

```python
# Visualize the distribution of injuries and deaths
injury_cols = ['NUMBER OF PERSONS INJURED', 'NUMBER OF PEDESTRIANS INJURED',
               'NUMBER OF CYCLIST INJURED', 'NUMBER OF MOTORIST INJURED']
death_cols = ['NUMBER OF PERSONS KILLED', 'NUMBER OF PEDESTRIANS KILLED',
              'NUMBER OF CYCLIST KILLED', 'NUMBER OF MOTORIST KILLED']

plt.figure(figsize=(10, 6))
data[injury_cols].sum().plot(kind='bar', color='y', label='Injured')
data[death_cols].sum().plot(kind='bar', color='r', label='Killed')
plt.title('Total Injuries and Deaths')
plt.xlabel('Type')
plt.ylabel('Count')
plt.legend()
plt.show()


```

```python
# Visualize the most common contributing factors for crashes
plt.figure(figsize=(10, 6))
top_factors = data['CONTRIBUTING FACTOR VEHICLE 1'].value_counts().head(10)
sns.barplot(x=top_factors.values, y=top_factors.index, color='b')
plt.title('Top Contributing Factors for Crashes')
plt.xlabel('Count')
plt.ylabel('Contributing Factor')
plt.show()
```

```python

# Plot the crashes on a scatter plot
plt.figure(figsize=(10, 8))
plt.scatter(df['LONGITUDE'], df['LATITUDE'], c='red', alpha=0.5, s=20)
plt.title('Crash Locations in New York City')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.show()


```

```python


# Visualize the distribution of crashes by time of day
data['CRASH TIME'] = pd.to_datetime(data['CRASH TIME'])
data['HOUR'] = data['CRASH TIME'].dt.hour

plt.figure(figsize=(10, 6))
sns.histplot(data['HOUR'], bins=24, kde=True)
plt.title('Distribution of Crashes by Time of Day')
plt.xlabel('Hour')
plt.ylabel('Count')
plt.xticks(range(24))
plt.show()

```

```python

# Visualize the top vehicle types involved in crashes
top_vehicle_types = data['VEHICLE TYPE CODE 1'].value_counts().head(10)

plt.figure(figsize=(10, 6))
sns.barplot(x=top_vehicle_types.values, y=top_vehicle_types.index, color='b')
plt.title('Top Vehicle Types Involved in Crashes')
plt.xlabel('Count')
plt.ylabel('Vehicle Type')
plt.show()


```

```python
# Visualize the correlation matrix of numeric columns
numeric_columns = ['NUMBER OF PERSONS INJURED', 'NUMBER OF PERSONS KILLED',
                   'NUMBER OF PEDESTRIANS INJURED', 'NUMBER OF PEDESTRIANS KILLED',
                   'NUMBER OF CYCLIST INJURED', 'NUMBER OF CYCLIST KILLED',
                   'NUMBER OF MOTORIST INJURED', 'NUMBER OF MOTORIST KILLED']

numeric_data = data[numeric_columns]

plt.figure(figsize=(10, 8))
sns.heatmap(numeric_data.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Matrix of Numeric Columns')
plt.show()

```
